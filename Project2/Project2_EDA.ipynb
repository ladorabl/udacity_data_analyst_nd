```python
# s10
import pandas as pd
# this prepares a new dataframe subset such that I drop all columns that contain 'player' in the column name string
cols = [c for c in match_detail.columns if 'player' not in c]

# this will be a list of dataframes, with each dataframe containing all matches that a player has play
player_matches = []
# function to generate player_match table
def generatePlayerMatchInfo():
    # iterate through each player in player table and search for player in home_player_N / away_player_N columns of match table
    for row in  player.itertuples():
        this_player_api_id =  row.player_api_id
        
        # create sub-dataframe of all matches this player_api_id is in by searching home_player columns. Drop player columns afterwards
        home_matches = match_detail [match_detail[['home_player_1' ,'home_player_2' ,'home_player_3' ,'home_player_4' ,'home_player_5' ,'home_player_6' ,'home_player_7' ,'home_player_8' ,'home_player_9' ,'home_player_10' 
                                            ,'home_player_11']].eq(this_player_api_id).any(1)][cols]
        # for each home match, declare player_team column based on home_team column
        home_matches['player_team'] = home_matches['home_team']
        home_matches['player_home_or_away']  = 'home'
        
        # do the same for away matches as was done for home matches
        away_matches = match_detail [match_detail[['away_player_1' ,'away_player_2' ,'away_player_3' ,'away_player_4' ,'away_player_5' ,'away_player_6' ,'away_player_7' ,'away_player_8' ,'away_player_9'
                                                   ,'away_player_10' ,'away_player_11']].eq(this_player_api_id).any(1)][cols]
        away_matches['player_team'] = away_matches['away_team']
        away_matches['player_home_or_away']  = 'away'
        
        # concatenate player's home matches and away matches
        this_player_match = pd.concat([home_matches, away_matches], ignore_index=True)
        # change this_plaer_match 'id' to 'match_id', so that I can prep for merch of player dataframe and player_match_data frame
        this_player_match.rename(columns={'id' : 'match_id'}, inplace=True)
        # make sure to save player_api_id for these rows
        this_player_match['player_api_id'] = this_player_api_id
        # merge player dataframe with player_match dataframe on player_api_id
        this_player_match = player[ player.player_api_id ==  this_player_api_id].merge(this_player_match, how='outer', on='player_api_id')
        # rename id to player_id, since id came from player table
        this_player_match.rename(columns={'id' : 'player_id'}, inplace=True)
        
        # add this result to player_matches array list. This will be concatenated later
        player_matches.append(this_player_match)
generatePlayerMatchInfo()
```

```python
# s15 compute
from lxml import etree

pm_gb_match_id = player_match.groupby('match_id')


def generate_dict_from_xpath(xpath_results):
    this_dict = {}
    for result in xpath_results:
        xpath_result = result.text
        if not this_dict:
            this_dict[xpath_result] = 1
        elif  xpath_result not in this_dict :
            this_dict[xpath_result] = 1
        else : 
            this_dict[xpath_result] = this_dict[xpath_result] + 1
    return this_dict
    

def compute_goal_scorers_and_assistants(this_match):
    this_goal_xml = this_match.goal
    
    if type(this_goal_xml) == str:
        root = etree.fromstring(this_goal_xml)

        player_goals_xpath_results = root.xpath('/goal/value/player1')
        player_assists_xpath_results = root.xpath('/goal/value/player2')

        # for each match, get player_api_id list of goal scorer
        player_scores_dict = generate_dict_from_xpath(player_goals_xpath_results)

        # for each match, get player_api_id list of assist maker 
        player_assists_dict = generate_dict_from_xpath(player_assists_xpath_results)

        # for this match_id, update each player's scores and assists from the dicts
        player_match_group_from_match_id = pm_gb_match_id.get_group(this_match.id)

        for player_api_id in player_scores_dict:
            goal_count = player_scores_dict[player_api_id]

            player_match_index =  player_match_group_from_match_id[player_match_group_from_match_id.player_api_id == int(player_api_id)].index
            player_match.iloc[player_match_index, player_match.columns.get_loc('player_goal_count')] = goal_count

        for player_api_id in player_assists_dict:
    
            assist_count = player_assists_dict[player_api_id]
            player_match_index =  player_match_group_from_match_id[player_match_group_from_match_id.player_api_id == int(player_api_id)].index
            player_match.iloc[player_match_index, player_match.columns.get_loc('player_assist_count')] = assist_count

#compute_goal_scorers_and_assistants(  match_detail [match_detail.id ==  player_match_ex1.match_id.item()])

def compute_foul_committers_and_receivers(this_match):
    this_match_foul = this_match.foulcommit.item()
    
    if type(this_match_foul) == str:
        root = etree.fromstring(this_match_foul)
        
        player_committed_foul_xpath_results = root.xpath('/foulcommit/value/player1')
        player_received_foul_xpath_results = root.xpath('/foulcommit/value/player2')
        
        players_committed_foul_dict = generate_dict_from_xpath(player_committed_foul_xpath_results)
        players_receivers_of_foul_dict = generate_dict_from_xpath(player_received_foul_xpath_results)
        
        player_match_group_from_match_id = pm_gb_match_id.get_group(this_match.id.item())
    
        for player_api_id in players_committed_foul_dict:
            foulcommit_count = players_committed_foul_dict[player_api_id]
            

            player_match_index =  player_match_group_from_match_id[player_match_group_from_match_id.player_api_id == int(player_api_id)].index
            player_match.iloc[player_match_index, player_match.columns.get_loc('player_foulcommit_count')] = foulcommit_count

        for player_api_id in players_receivers_of_foul_dict:
    
            foulreceived_count = players_receivers_of_foul_dict[player_api_id]
            player_match_index =  player_match_group_from_match_id[player_match_group_from_match_id.player_api_id == int(player_api_id)].index
            player_match.iloc[player_match_index, player_match.columns.get_loc('player_foulreceived_count')] = foulreceived_count

    

#player_match['player_foulreceived_count'] = None
#player_match['player_foulcommit_count'] = None

#compute_goal_scorers_and_assistants(  match_detail [match_detail.id ==  player_match_ex1.match_id.item()])
compute_foul_committers_and_receivers(match_detail[match_detail.id == 4768])
#root = etree.fromstring(match_detail[match_detail.id == 4768].foulcommit.item())
```
