```python
def getMatchWinner(row):
    if row['home_team_goal'] > row['away_team_goal']:
        return row['home_team']
    elif row['away_team_goal'] > row['home_team_goal']:
        return row['away_team']
    else:
        return 'Tie'
#belgium_2013_2014_matches['Winner'] =  belgium_2013_2014_matches.apply(lambda x : getMatchWinner(x), axis=1)
match_detail['Winner'] = match_detail.apply(lambda x : getMatchWinner(x), axis=1)
```
